```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pyEXP
import EXPtools
#plt.style.use("~/Work/research/MW_viz/matplotlib.mplstyle")
```

```python
# Load coefficients 

npart = 100000
HHalo = EXPtools.utils.halo.ICHernquist(size=npart)
```

```python
SphericalH = HHalo.triaxial(axis_ratios=[1, 1, 1], rot_angle=0.0, rot_axis='y')
      
```

```python
dipoles_power = np.zeros((10, 10))
monopoles_power = np.zeros((10, 10))
#dipoles_m0 = np.zeros((10, 10))
#dipoles_m1 = np.zeros((10, 10))
#dipoles_m_1 = np.zeros((10, 10))
p1 = np.zeros((10, 10))
p3 = np.zeros((10, 10))
p4 = np.zeros((10, 10))
#a = 1
barr = np.linspace(0.5, 1, 10)
carr = np.linspace(0.5, 1, 10)
ca = np.zeros(10)
ba = np.zeros(10)
i=0
for c in range(len(carr)):
    print(c)
    for b in range(c, len(barr)):
        SphericalH = HHalo.triaxial(axis_ratios=[1, barr[b], carr[c]], rot_angle=0.0, rot_axis='y')
        basis, coefs = EXPtools.basis_builder.makebasis(SphericalH, np.ones(npart)/npart, basis_model='Hernquist',
                                                basis_id='sphereSL', time=i, nbins=50, rmin=0.02, rmax=2,
                                                log_space=True,
                                                lmax=5, nmax=10, scale=1, modelname='SLGrid.empirical_hern_tri_halo', 
                                                cachename=".slgrid_sph_hern_triaxial", 
                                                add_coef=True, coef_file='hern_triaxial_coef.h5')
        i+=1
        dipoles_power[b][c] = coefs.Power()[0][2]
        monopoles_power[b][c] = coefs.Power()[0][0]
        p1[b][c] = coefs.Power()[0][1]
        p3[b][c] = coefs.Power()[0][3]
        p4[b][c] = coefs.Power()[0][4]
```

```python
fig, ax = plt.subplots(1, 1)
im = plt.imshow((dipoles_power/monopoles_power).T, origin='lower', cmap='magma_r',
                vmin=0, vmax=0.02, extent=[0.2, 1, 0.2, 1])
cbar = plt.colorbar(im)
cbar.set_label(r'$\rm{P_{l=2}/P_{l=0}}$')
ax.set_xlabel(r'$b/a$')
ax.set_ylabel(r'$c/a$')
ax.text(0.45, 0.6, r'$\rm{Prolate}$', rotation=45, fontsize=15)
ax.text(0.8, 0.1, r'$\rm{Oblate}$', fontsize=15)
ax.text(0.8, 1.03, r'$\rm{Spherical}$', fontsize=15)
ax.set_facecolor('lightgrey')
#plt.savefig('../figures/shape_power_relationship.pdf', bbox_inches='tight')
```
